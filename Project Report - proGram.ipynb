### Data Extraction and Cleaning 

Since, the datasets are very big and cannot be loaded completely in python with ease, we have taken a random sample of 100,000 Instagram Profiles from the profiles file. Furthermore, we have filtered the Instagram Posts dataset by mapping the data using profile id with the sampled out profiles. 


#### Process of Data Extraction and Cleaning: 

1. Instagram Profiles

   - To clean data from the sampled 100,000 dataset of profiles, we first dropp the cts column which is not required to determine the quality score. 
   - Then, we drop profiles which have null values in the columns - sid, profile_id and profile_name. These columns are important for mapping to the other files and help us identify the user. 
   - Since, our requirement focuses on catering to only mini and micro influencers on instagram, which is determined by the range of their followers, we filter out these influencer profiles with the help of column "followers". We will only take profiles which have followers in the range of 1000 to 100000.
   
   
   
2. Locations

    To clean data of the entire dataset of locations, we will drop the locations which have null values in the columns sid, id, name and cd (country code). These columns are important for mapping to other files and help us identify the location. 

```python
import pandas as pd
import numpy as np
import random
import re
from datetime import datetime
from dateutil.relativedelta import relativedelta
```

```python
## Import data for instagram profiles file and sample random 100,000 profiles

filename_profiles = r"proGram_src\instagram_profiles.csv"

# Get total number of profiles in the file
total_num_profiles = sum(1 for line in open(filename_profiles, encoding="utf8")) - 1 # Total rows in original file
print (f"Total number of profiles in the dataset: {total_num_profiles}")

# Randomly skip rows from file to generate a sample of 100,000 profiles
random.seed(33)
sample_size = 100000                                                                 # Desired sample size
skip = sorted(random.sample(range(1,total_num_profiles+1), total_num_profiles-sample_size))
profiles_sample = pd.read_csv(filename_profiles, skiprows=skip, sep='\t')

# Get total number of profile sampled
print (f"Total number of profiles sampled: {profiles_sample.shape[0]}")

# Display fetched dataset
profiles_sample.head(10)
```

```python
## Clean data for instagram profiles 

# Drop column cts as it is not required
profiles_sample.drop(columns=['cts'], axis=1, inplace=True)

# Filter rows where sid, profile_id and profile_name is not null
profiles_sample = profiles_sample.dropna(subset=["sid", "profile_id", "profile_name"])

# Filter mino and micro influencer profiles (min followers=1000 and max followers=100000)
min_followers = 1000
max_followers = 100000
profiles_sample = profiles_sample[(profiles_sample.followers > min_followers) & (profiles_sample.followers < max_followers)]

# Convert profile_id column to integer64 
profiles_sample["profile_id"] = profiles_sample["profile_id"].astype('int64')

# Convert followers, followings and no of posts to integer
profiles_sample["following"] = profiles_sample["following"].astype(int)
profiles_sample["followers"] = profiles_sample["followers"].astype(int)
profiles_sample["n_posts"] = profiles_sample["n_posts"].astype(int)

# Get total number of profiles after cleaning
no_of_profiles_sample = profiles_sample.shape[0]
print (f"Number of profiles after cleaning: {no_of_profiles_sample}")

# Display final sample of profiles after cleaning
profiles_sample.head(10)
```

```python
## Import data from instagram locations file 

filename_locations = r"proGram_src\instagram_locations.csv"

# Create list of columns required for import
columns_req = ['sid', 'id', 'name', 'street', 'zip', 'city', 'region', 'cd', 'dir_city_name', 'dir_city_slug', 'dir_country_id', 'dir_country_name', 'lat', 'lng']

# Load data
locations_dataset = pd.read_csv(filename_locations, usecols=columns_req, sep='\t')

# Get total number of locations in the file
print (f"Total number of locations in dataset: {locations_dataset.shape[0]}")

# Display fetched dataset
locations_dataset.head(10)
```

```python
## Clean data for instagram locations

# Filter rows where sid, id, name, and country (cd) is not null
locations_sample = locations_dataset.dropna(subset=["sid", "id", "name", "cd"])

# Convert profile_id column to integer64 
locations_sample["id"] = locations_sample["id"].astype('int64')

# Get total number of locations after cleaning
no_of_locations_sample = locations_sample.shape[0]
print (f"Number of locations after cleaning: {no_of_locations_sample}")

# Display final sample of locations after cleaning
locations_sample.head(10)
```

```python
## Import data from instagram posts file 

filename_posts = r"proGram_src\instagram_posts.csv"

# Sample posts dataset by filtering out posts based on sample profiles by matching profile ids
chunksize = 100000       # we will import file in chunks as data is too huge
i = 1
for chunk in pd.read_csv(filename_posts, sep='\t', low_memory=False, chunksize=chunksize):
        posts = chunk[chunk["profile_id"].isin(profiles_sample["profile_id"])]
        if i == 1:
            posts_sample = posts
        else:
            posts_sample = pd.concat([posts_sample, posts])
        i = i+1

# Get total number of posts fetched from the file based on profiles sample
print (f"Total number of posts fetched from file based on profiles sample: {posts_sample.shape[0]}")

# Display fetched posts
posts_sample.head(10)
```

```python
## Clean data for instagram posts

# Filter rows where sid is not null
posts_sample = posts_sample.dropna(subset=["sid"])

# Convert profile_id and location_id columns to integer64 
posts_sample["profile_id"] = posts_sample["profile_id"].astype('Int64')
posts_sample["location_id"] = posts_sample["location_id"].fillna(0)
posts_sample["location_id"] = posts_sample["location_id"].astype('Int64')

# Convert number of likes and number of comments columns to integer
posts_sample["numbr_likes"] = posts_sample["numbr_likes"].astype(int)
posts_sample["number_comments"] = posts_sample["number_comments"].astype(int)

# Convert cts to date time format
posts_sample['cts'] = pd.to_datetime(posts_sample['cts'])

# Get total number of posts after cleaning
no_of_posts_sample = posts_sample.shape[0]
print (f"Number of posts after cleaning: {no_of_posts_sample}")

# Display final sample of posts after cleaning
posts_sample.head(10)
```

```python
keyword = 'fashion'
```

```python
## Metrics 1: Total Number of Followers
## Metrics 2: Total Number of Posts (Based on profiles data)

df_quality_score = pd.DataFrame(profiles_sample[["profile_id", "followers", "n_posts"]])
df_quality_score.set_index('profile_id', inplace=True)

# Get Total number of posts of user by counting number of rows for each user in posts sample
# Note: This is different than Total number of posts of user in profiles sample 
posts_by_users = posts_sample.groupby(by=['profile_id'])['sid'].count().rename('total_posts', inplace=True)
df_quality_score = df_quality_score.merge(posts_by_users, left_index=True, right_index=True)


df_quality_score.head(10)
```

```python
## Metrics 3: Engagement Rate of each profile/user


## Engagement Score = Number of likes + Number of comments (of all posts of the user) / Total number of followers of user

# Get Total number of likes & comments by summation of all numbr_likes and number_comments columns for each user 
sum_cols = ["numbr_likes", "number_comments"]
posts_by_users = posts_sample.groupby(by=['profile_id'])[sum_cols].sum()
df_quality_score = df_quality_score.merge(posts_by_users, left_index=True, right_index=True)


# Calculate Engagement Score for all users
df_quality_score["eng_score"] = ((df_quality_score['numbr_likes'] + df_quality_score['number_comments'])
                                 /df_quality_score['followers']).round(2)


# Display Engagament metrics
df_quality_score.head(10)
```

```python
## Metrics 4: Recent Engagement Rate of each profile/user

## Calculating Engagement Score and Rate for recent 10 posts of the user
## If last post was made before 2018 --> Recent Engagement Score = 0

# Get most recent 10 posts for each user and create new posts sample
posts_sample_temp = posts_sample.copy()
posts_sample_temp.sort_values(by=['profile_id','cts'], ascending=[True,False], inplace=True)
posts_sample_temp = posts_sample_temp.groupby('profile_id').head(10)


## For the most recent 10 posts we will calculate:
## Recent Engagement Score = Number of likes + Number of comments (of all posts of the user) / Total number of followers of user

# Get Total number of likes & comments by summation of all numbr_likes and number_comments columns for each user 
sum_cols = ["numbr_likes", "number_comments"]
posts_by_users = posts_sample_temp.groupby(by=['profile_id'])[sum_cols].sum()
df_quality_score = df_quality_score.merge(posts_by_users, left_index=True, right_index=True, suffixes=['', '_rec'])

# Calculate Engagement Score for all users
df_quality_score["eng_score_rec"] = ((df_quality_score['numbr_likes_rec'] + df_quality_score['number_comments_rec'])
                                     /df_quality_score['followers']).round(2)


## Check if last post was made before 2018
year_threshold = 2018

# Create dummy variable for users whose recent post was before 2018 (value=0) or not (value=1)
posts_sample_temp = posts_sample_temp.groupby('profile_id').head(1)
posts_sample_temp['recent_post_before_2018'] = posts_sample_temp['cts'].map(lambda x: 0 if x.year < year_threshold else 1)
df_quality_score = df_quality_score.merge(posts_sample_temp[['profile_id', 'recent_post_before_2018']], 
                                          left_index=True, right_on='profile_id').set_index('profile_id')

# Make Engagement Score = 0 if user's recent post was before 2018 
df_quality_score["eng_score_rec"] = df_quality_score["eng_score_rec"] * df_quality_score['recent_post_before_2018']
    

# Display Recent Engagement metrics
df_quality_score.head(10)
```

```python
## Metrics 5: Average percentage of increase in engagement score (Growth)

# Get a copy of posts samples 
posts_sample_temp = posts_sample.copy()

## For every post we will calculate:
## Engagement Score = Number of likes + Number of comments (of all posts of the user) / Total number of followers of user

# Get Total number of followers of user in posts sample dataset
posts_sample_temp = posts_sample_temp.merge(profiles_sample[["profile_id", "followers"]], left_on="profile_id", right_on="profile_id", how="left")

# Get Engagement rate for each post
posts_sample_temp["eng_score"] = ((posts_sample_temp["numbr_likes"] + posts_sample_temp["number_comments"]) 
                                  / posts_sample_temp["followers"])

# Sort posts by users and timestamp (recent first)
posts_sample_temp.sort_values(by=['profile_id','cts'], ascending=[True,False], inplace=True)

# Get percentage of increase in engagement score for consecutive posts for user
posts_sample_temp["perc_inc_eng_score"] = posts_sample_temp.groupby('profile_id')["eng_score"].pct_change() * 100

# To handle np.inf values, we will make np.inf or -np.inf as np.nan (not to be considered for average perc increase)
posts_sample_temp["perc_inc_eng_score"] = posts_sample_temp["perc_inc_eng_score"].replace([np.inf, -np.inf], np.nan)

# Get average of percentage of increase in engagement score for all posts for each user
posts_sample_temp = posts_sample_temp.groupby('profile_id')["perc_inc_eng_score"].mean().round(2)

# Consider average of percentage increase as 0 if Nan
posts_sample_temp = posts_sample_temp.fillna(0)

# If average of percentage of increase in eng score is negative then we will consider this score as 0
posts_sample_temp = posts_sample_temp.mask((posts_sample_temp < 0), 0)
df_quality_score = df_quality_score.merge(posts_sample_temp, left_index=True, right_index=True)

# Display metrics
df_quality_score.head(20)
```

```python
## Metrics 6: Average no. of likes per post by Total Followers
## Metrics 7: Average no. of comments per post by Total Followers

# Get Average number of likes i.e. (total numbr_likes / total posts for each user) / total followers
df_quality_score['avg_likes_by_foll'] = ((df_quality_score['numbr_likes'] / df_quality_score['total_posts'])/df_quality_score['followers']).round(2)

# Get Average number of comments i.e. (total comments_likes / total posts for each user) / total followers 
df_quality_score['avg_com_by_foll'] = ((df_quality_score['number_comments'] / df_quality_score['total_posts'])/df_quality_score['followers']).round(2)
                         
# Display metrics
df_quality_score.head(20)
```

```python
## Metrics 8: Hashtag Score - Ideal number of Hashtags per post 

## Get number of hashtags used by each user in their posts 

# Import file to map number of hashtags vs score to be given to the user 
filename_hashtags_score = r"proGram_src\NumberofHashtags_Score.csv"
hashtags_score_dataset = pd.read_csv(filename_hashtags_score, sep=',')

# Get a copy of posts samples with profile_id and description (description is the caption)
posts_sample_temp = posts_sample[['profile_id', 'description']].copy()

# Pattern for matching hashtag
pattern = "[#][\S]*"

# Create new variable to store average number of hashtags per post used by user 
posts_sample_temp["no_of_hashtags"] = posts_sample_temp["description"].str.count(pattern, flags=re.IGNORECASE)
posts_sample_temp["hashtag_score"] = 0
for row in hashtags_score_dataset.iterrows():
    posts_sample_temp["hashtag_score"] = np.where(posts_sample_temp["no_of_hashtags"].between(row[1]["Min"],row[1]["Max"]), row[1]["Score"], posts_sample_temp["hashtag_score"])


# Calculate average of hashtag score for each user
posts_sample_temp = posts_sample_temp.groupby('profile_id').mean().round(2)
df_quality_score = df_quality_score.merge(posts_sample_temp['hashtag_score'], left_index=True, right_index=True)

# Display metrics
df_quality_score.head(10)
```

```python
## Metrics 9: Keyword match - Profile Bio

## Get users whose bios have the keyword
    
# Get a copy of profiles samples with profile_id and description (description is the bio)
profiles_sample_temp = profiles_sample[['profile_id', 'description']].copy().set_index('profile_id')
    
# Create new variable to mark if user has bio with the keyword (Yes="1", No="0")
profiles_sample_temp["bio_with_keyword"] = profiles_sample_temp["description"].str.contains(keyword, flags=re.IGNORECASE)
profiles_sample_temp["bio_with_keyword"].fillna(False, inplace=True) # If bio is null, then consider match as False
profiles_sample_temp["bio_with_keyword"] = profiles_sample_temp["bio_with_keyword"].astype(int)
df_quality_score = df_quality_score.merge(profiles_sample_temp['bio_with_keyword'], left_index=True, right_index=True)
    
# Get total number of profiles with keyword in their bios
print(f"Number of profiles with keyword: {keyword} in their bios = {df_quality_score['bio_with_keyword'].sum()}")

# Display metrics
df_quality_score.head(10)
```

```python
## Metrics 10: Keyword match (number of posts) - in post captions 
## Metrics 11: Keyword match (number of posts/total posts) - in posts captions

## Get number of posts that have the keyword for each user

# Get a copy of posts samples with profile_id and description (description is the caption)
posts_sample_temp = posts_sample[['profile_id', 'description']].copy()
    
# Create new variable to store number of posts that have the keyword for each user
posts_sample_temp["posts_with_keyword"] = posts_sample_temp["description"].str.contains(keyword, flags=re.IGNORECASE)
posts_sample_temp["posts_with_keyword"].fillna(False, inplace=True) # If caption is null, then consider match as 0
posts_sample_temp = posts_sample_temp.groupby('profile_id').sum()
posts_sample_temp["posts_with_keyword"] = posts_sample_temp["posts_with_keyword"].astype(int)
df_quality_score = df_quality_score.merge(posts_sample_temp['posts_with_keyword'], left_index=True, right_index=True)

## Get number of posts that have the keyword/total posts for each user
df_quality_score["posts_with_keyword_by_total"] = (df_quality_score['posts_with_keyword'] / df_quality_score['total_posts']).round(2)

# Display metrics
df_quality_score.head(10)
```

```python
## Metrics 12: Keyword match - Hashtags

## Get number of hashtags used by each user in their posts that have the keyword 

# Get a copy of posts samples with profile_id and description (description is the caption)
posts_sample_temp = posts_sample[['profile_id', 'description']].copy()

# Pattern for matching hashtag
pattern = "[#][\S]*"+keyword+"[\S]*"

# Create new variable to store number of hashtags used by user 
posts_sample_temp["hashtag_with_keyword"] = posts_sample_temp["description"].str.count(pattern, flags=re.IGNORECASE)
posts_sample_temp = posts_sample_temp.groupby('profile_id').sum()
posts_sample_temp["hashtag_with_keyword"] = posts_sample_temp["hashtag_with_keyword"].astype(int)
df_quality_score = df_quality_score.merge(posts_sample_temp['hashtag_with_keyword'], left_index=True, right_index=True)

# Display metrics
df_quality_score.head(10)
```

```python
## Metrics 13: Requirement Specific Engagement Rate - only for posts where keyword matches

## Get posts that have the keyword for each user

# Get a copy of posts samples with profile_id and description (description is the caption)
posts_sample_temp = posts_sample.copy()
    
# Create new variable to store number of posts that have the keyword for each user
posts_sample_temp["posts_with_keyword"] = posts_sample_temp["description"].str.contains(keyword, flags=re.IGNORECASE)
posts_sample_temp["posts_with_keyword"].fillna(False, inplace=True) # If caption is null, then consider match as 0

# Only take posts which have the keyword
posts_sample_temp = posts_sample_temp[posts_sample_temp['posts_with_keyword'] == True]

# Get Total number of likes & comments by summation of all numbr_likes and number_comments columns for each user 
sum_cols = ["numbr_likes", "number_comments"]
posts_by_users = posts_sample_temp.groupby(by=['profile_id'])[sum_cols].sum()
df_quality_score = df_quality_score.merge(posts_by_users, left_index=True, right_index=True, suffixes=['', '_keyword'], how="left")
df_quality_score["numbr_likes_keyword"].fillna(0, inplace=True)
df_quality_score["number_comments_keyword"].fillna(0, inplace=True)

# Calculate Engagement Score for all users for posts with keyword
df_quality_score["eng_score_keyword"] = ((df_quality_score['numbr_likes_keyword'] + df_quality_score['number_comments_keyword'])
                                         /df_quality_score['followers']).round(2)


# Display metrics
df_quality_score.head(10)
```

```python
### Calculate Overall Quality Score

# Import file to map metrics vs weightage given to it to calculate quality score
filename_metrics_weightage = r"proGram_src\Metrics_Weightage.xlsx"
metrics_weightage = pd.read_excel(filename_metrics_weightage, index_col="Metrics")

metrics_weightage.head(20)
```

```python
# Convert Metrics column to list
metrics_list = metrics_weightage.index.to_list()
print(f"Metrics: {metrics_list}")

# Convert Weights column to list
weights_list = metrics_weightage["Weights"].to_list()
print(f"Weights: {weights_list}")

# Grade every user for each metrics (out of 100) and create new dataframe of graded metrics
# Grade = User's value of metrics / Max possible value of metrics * 100
df_quality_score_grade = ((df_quality_score.loc[:, metrics_list] / df_quality_score.loc[:, metrics_list].max()) * 100).round(2)

# Calculate Overall Quality Score by multiplying weightage of each metrics to the metrics grade 
df_quality_score_grade["overall_score"] = df_quality_score_grade.mul(weights_list).sum(axis=1)

# Display Overall Quality Score
df_quality_score_grade.head(20)
```

```python
# Understand the variation in overall quality score 
df_quality_score_grade["overall_score"].describe()
```

```python
## Calculate Quality score percentage 

# Get max possible value of overall quality score
max_overall_quality_score = df_quality_score_grade["overall_score"].max()
print(f"Maximum overall quality score is: {max_overall_quality_score}")

# Get quality score percentage for all users (considering max value as total points for quality score)
df_quality_score_grade["quality_score (in %)"] = ((df_quality_score_grade["overall_score"] / max_overall_quality_score) * 100).round(2)

# Display quality score percentage
df_quality_score_grade.head(20)
```

```python
## Calculate Quality Score percentile rank of overall quality score

df_quality_score_grade['quality_score (percentile_rank)'] = df_quality_score_grade["overall_score"].rank(pct = True).round(2)

# Display Quality score percentile rank
df_quality_score_grade.head(20)
```

```python
## Get Top 10 profiles with the best quality score
df_quality_score_grade.sort_values(by=['overall_score'], ascending=False, inplace=True)

df_quality_score_grade.head(10)
```
